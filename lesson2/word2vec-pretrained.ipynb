## Word2vec - предобученная модель

```python
import gensim
```

```python
#скачиваем с https://rusvectores.org/ru/models/
#модели .gz можно ипользовать с флагом binary=True

model_file="ruscorpora_upos_skipgram_300_5_2018.vec"
model = gensim.models.KeyedVectors.load_word2vec_format(model_file, binary=False)
```

```python
model['дом']
```

```python
[x for x in model.index_to_key if x.startswith('дом')][:5]
```

```python
model['дом_NOUN']
```

```python
len(model['дом_NOUN'])
```

```python
model.most_similar(positive=["любить_VERB"])
```

```python
model.most_similar(positive=["папа_NOUN", "брат_NOUN"], negative=["мама_NOUN"])
```

```python
model.similarity("плохой_ADJ", "герой_NOUN")
```

```python
model.similarity("хороший_ADJ", "герой_NOUN")
```
