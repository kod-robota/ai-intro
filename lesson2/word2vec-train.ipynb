## word2vec - обучение модели 

```python
from gensim.models import Word2Vec
```

```python
# скачиваем TXT с https://all-the-books.ru/books/tolstoy-lev-anna-karenina/?ysclid=l5i18yi39b593772545

text_file = "anna-karenina.txt"
```

```python
import unicodedata
import sys

tbl = dict.fromkeys(i for i in range(sys.maxunicode)
                      if unicodedata.category(chr(i)).startswith('P'))

def remove_punctuation(text):
    return text.translate(tbl)
```

```python
data=[]
with open(text_file, "r") as f:
    for line in f:
        line = line.strip().lower()
        #print(line)
        out = remove_punctuation(line)
        #print(out)
        line_data = out.split()
        #print(line_data)
        if line_data:
            data.append(line_data)
```

```python
len(data)
```

```python
data[:10]
```

```python
model = Word2Vec(sentences=data, vector_size=100, window=5, min_count=1, workers=4)
```

```python
model.build_vocab(data)
```

```python
len(model.wv.index_to_key)
```

```python
model.wv.index_to_key[:5]
```

```python
[x for x in model.wv.index_to_key if x.startswith('дом')][:5]
```

```python
model.train(data, total_examples=model.corpus_count, epochs=100, report_delay=1)
```
